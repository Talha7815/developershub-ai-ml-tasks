## 1. Import Required Libraries
First, let's import all the Python libraries we'll need for data analysis and visualization.

```python
# Install required libraries (if not already installed)
# !pip install pandas matplotlib seaborn scikit-learn

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization style for better looking plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

print("✅ Libraries imported successfully!")
```

## 2. Load the Iris Dataset
The Iris dataset is a classic dataset in machine learning. It contains measurements of 150 iris flowers from three different species.

```python
# Load the dataset using seaborn
iris = sns.load_dataset('iris')
print("✅ Dataset loaded successfully!")

# Display the first few rows to understand the structure
print("\n🔍 First 5 rows of the dataset:")
iris.head()
```

### Initial Observations from the First Few Rows:
- The dataset has 5 columns: 4 features (sepal_length, sepal_width, petal_length, petal_width) and 1 target variable (species)
- All measurements appear to be in centimeters
- The first 3 rows show Setosa species with varying measurements

## 3. Dataset Information and Structure
Let's examine the dataset's structure, data types, and check for any missing values.

```python
# Check the shape of the dataset
print("📊 Dataset Shape:", iris.shape)
print("   - Rows:", iris.shape[0])
print("   - Columns:", iris.shape[1])

# Check column names
print("\n📋 Column Names:")
print(iris.columns.tolist())

# Get detailed info about the dataset
print("\nℹ️ Dataset Info:")
print(iris.info())

# Check for missing values
print("\n🔍 Missing Values in Each Column:")
print(iris.isnull().sum())
```

### Key Findings from Dataset Info:
- ✅ **No missing values** - The dataset is clean and ready for analysis
- ✅ All features are numerical (float type) - Perfect for visualization
- ✅ Species column is categorical (object type)
- Dataset has exactly **150 samples** - A good size for exploration

## 4. Summary Statistics
Let's generate statistical summaries to understand the distribution of each feature.

```python
# Generate summary statistics
print("📊 Summary Statistics:")
print(iris.describe())

# Check the distribution of target classes
print("\n🌸 Species Distribution:")
print(iris['species'].value_counts())
```

### Insights from Summary Statistics:

| Feature | Mean | Min | Max | Range |
|---------|------|-----|-----|-------|
| Sepal Length | 5.84 cm | 4.3 cm | 7.9 cm | 3.6 cm |
| Sepal Width | 3.06 cm | 2.0 cm | 4.4 cm | 2.4 cm |
| Petal Length | 3.76 cm | 1.0 cm | 6.9 cm | 5.9 cm |
| Petal Width | 1.20 cm | 0.1 cm | 2.5 cm | 2.4 cm |

**Key Observations:**
- Petal length has the widest range (5.9 cm), showing high variability
- Dataset is perfectly balanced with 50 samples of each species
- Sepal measurements are more conservative in range compared to petal measurements

## 5. Data Visualization
Now let's create visualizations to better understand the relationships and patterns in the data.

### 5.1 Scatter Plots - Showing Relationships Between Features
Scatter plots help us see how two features relate to each other and how different species cluster together.

```python
# Scatter plot 1: Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', 
                hue='species', style='species', s=100, palette='deep')
plt.title('Figure 1: Sepal Length vs Sepal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### Observations from Sepal Scatter Plot:
- **Setosa** (blue) shows distinct clustering - generally has smaller sepal length but larger sepal width
- **Versicolor** (orange) and **Virginica** (green) overlap significantly in this view
- Sepal features alone don't perfectly separate all three species
- There appears to be a weak negative correlation for some species

```python
# Scatter plot 2: Petal Length vs Petal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', 
                hue='species', style='species', s=100, palette='deep')
plt.title('Figure 2: Petal Length vs Petal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### Observations from Petal Scatter Plot:
- **Setosa** (blue) forms a completely separate cluster in the bottom-left
- Much clearer separation between species compared to sepal features
- **Strong positive correlation** between petal length and petal width
- Versicolor and Virginica show some overlap but are more distinguishable
- Petal features are clearly better for species identification

### 5.2 Pair Plot - All Feature Combinations
A pair plot shows relationships between all features simultaneously.

```python
# Create a pair plot
plt.figure(figsize=(12, 10))
pair_plot = sns.pairplot(iris, hue='species', diag_kind='hist', 
                         palette='deep', height=2.5)
pair_plot.fig.suptitle('Figure 3: Pair Plot of All Iris Features', 
                       y=1.02, fontsize=14, fontweight='bold')
plt.show()
```

#### Insights from Pair Plot:

| Feature Pair | Separation Quality | Notes |
|--------------|-------------------|-------|
| Petal Length vs Petal Width | ⭐⭐⭐ Excellent | Clear clusters, minimal overlap |
| Petal Length vs Sepal Length | ⭐⭐ Good | Some separation, some overlap |
| Sepal Length vs Sepal Width | ⭐ Poor | Significant overlap between species |
| Petal Width vs Sepal Width | ⭐⭐ Good | Moderate separation |

**The diagonal histograms** show the distribution of each feature:
- Petal features show **bimodal or trimodal** distributions (good for classification)
- Sepal features show more **overlapping normal distributions**

### 5.3 Histograms - Understanding Value Distributions
Histograms show how values are distributed for each feature.

```python
# Create histograms for all features
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
titles = ['Sepal Length Distribution', 'Sepal Width Distribution', 
          'Petal Length Distribution', 'Petal Width Distribution']
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon']

for i, (feature, title, color) in enumerate(zip(features, titles, colors)):
    row, col = i // 2, i % 2
    axes[row, col].hist(iris[feature], bins=15, color=color, edgecolor='black', alpha=0.7)
    axes[row, col].set_title(f'Figure 4.{i+1}: {title}', fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel(feature.replace('_', ' ').title() + ' (cm)')
    axes[row, col].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Histogram Observations:
- **Sepal Length**: Appears roughly normal, centered around 5.8 cm
- **Sepal Width**: Also roughly normal but with possible outlier on lower end
- **Petal Length**: **Clear bimodal distribution** - two distinct peaks indicating different species
- **Petal Width**: Also bimodal, confirming species differences

The bimodal nature of petal features is a strong indicator that these features will be excellent for classification.

```python
# Histograms by species (to see distribution per species)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    for species in iris['species'].unique():
        subset = iris[iris['species'] == species]
        axes[row, col].hist(subset[feature], bins=10, alpha=0.7, label=species)
    axes[row, col].set_title(f'Figure 5.{i+1}: {titles[i]} by Species', fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel(feature.replace('_', ' ').title() + ' (cm)')
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].legend()

plt.tight_layout()
plt.show()
```

#### Species-wise Distribution Insights:
- **Setosa** (blue) has completely separate ranges for petal features - no overlap!
- **Versicolor** (orange) sits in the middle range for all features
- **Virginica** (green) occupies the higher end of measurements
- This explains why Iris dataset is perfect for teaching classification

### 5.4 Box Plots - Identifying Outliers
Box plots help us spot outliers and understand the spread of data.

```python
# Box plots for each feature
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    sns.boxplot(data=iris, x='species', y=feature, ax=axes[row, col], palette='Set2')
    axes[row, col].set_title(f'Figure 6.{i+1}: Box Plot of {titles[i]}', fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel('Species')
    axes[row, col].set_ylabel(feature.replace('_', ' ').title() + ' (cm)')

plt.tight_layout()
plt.show()
```

#### Outlier Analysis from Box Plots:

| Feature | Outliers Detected | Notes |
|---------|-------------------|-------|
| Sepal Length | None | Clean distribution per species |
| Sepal Width | Yes - 4 potential outliers | One Setosa, one Versicolor, two Virginica |
| Petal Length | None | Perfect separation between species |
| Petal Width | None | Clean distributions |

**Conclusion:** The dataset is very clean with only a few minor outliers in sepal width. These are likely natural variations rather than errors.

```python
# Combined box plot for comparison
plt.figure(figsize=(12, 6))
iris_melted = pd.melt(iris, id_vars=['species'], value_vars=features, 
                      var_name='Feature', value_name='Value')
sns.boxplot(data=iris_melted, x='Feature', y='Value', hue='species', palette='Set2')
plt.title('Figure 7: Comparative Box Plots of All Features', fontsize=14, fontweight='bold')
plt.xlabel('Features')
plt.ylabel('Values (cm)')
plt.xticks(rotation=45)
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 5.5 Correlation Heatmap
A heatmap shows how features correlate with each other.

```python
# Calculate correlation matrix
plt.figure(figsize=(10, 8))
numerical_cols = iris.select_dtypes(include=[np.number]).columns
correlation_matrix = iris[numerical_cols].corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1, center=0)
plt.title('Figure 8: Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

#### Correlation Analysis:

| Feature Pair | Correlation | Strength |
|--------------|-------------|----------|
| Petal Length vs Petal Width | **0.96** | Very Strong Positive |
| Petal Length vs Sepal Length | **0.87** | Strong Positive |
| Petal Width vs Sepal Length | **0.82** | Strong Positive |
| Sepal Length vs Sepal Width | **-0.11** | Very Weak Negative |

**Key Insight:** Petal features are highly correlated with each other and with sepal length, but sepal width behaves independently. This explains why sepal width is less useful for classification.

### 5.6 Violin Plots - Distribution Shape
Violin plots combine box plots with density distributions.

```python
# Violin plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    sns.violinplot(data=iris, x='species', y=feature, ax=axes[row, col], 
                   palette='Set2', inner='quartile')
    axes[row, col].set_title(f'Figure 9.{i+1}: Violin Plot of {titles[i]}', 
                             fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel('Species')
    axes[row, col].set_ylabel(feature.replace('_', ' ').title() + ' (cm)')

plt.tight_layout()
plt.show()
```
