## 🌟 Overview

This project implements a **Mental Health Support Chatbot** fine-tuned on the EmpatheticDialogues dataset. The chatbot can recognize and respond appropriately to various emotional states including sadness, anxiety, loneliness, and depression, with built-in crisis detection for suicide-related keywords.

**Purpose**: To provide empathetic, non-judgmental emotional support through natural conversation.

**Target Users**: Anyone seeking emotional support, stress relief, or someone to talk to.

## ✨ Features

### 🤖 **Intelligent Conversations**
- Emotion recognition from user input
- Context-aware responses
- Conversation memory

### 🛡️ **Safety First**

## 🏗️ Technical Architecture

### Model Details
| Component | Specification |
|-----------|--------------|
| **Base Model** | GPT-Neo 125M (EleutherAI) |
| **Fine-tuning Data** | EmpatheticDialogues (Facebook AI) |
| **Framework** | Hugging Face Transformers v5.0.0 |
| **Training** | 2 epochs on T4 GPU (~2 hours) |
| **Parameters** | 125 million |

### Dataset Statistics
| Split | Examples |
|-------|----------|
| **Training** | 76,668 conversations |
| **Validation** | 6,318 conversations |
| **Test** | 5,701 conversations |
| **Emotion Categories** | 32 |

### Technology Stack
├── Python 3.10+ # Core language
├── PyTorch 2.0+ # Deep learning framework
├── Transformers 5.0.0 # Hugging Face library
├── Streamlit 1.28+ # Web interface
├── Datasets 2.14+ # Dataset handling
├── Google Colab # Training environment
└── ngrok/localtunnel # Deployment tunneling
